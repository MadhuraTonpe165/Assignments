```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load the Titanic dataset
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
data = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
## 1. What is the distribution of passenger ages on the Titanic?
#Plotting Age count graph
sns.histplot(x='Age', data=data, kde=True, color="green")
plt.title('Survival by age')
plt.show()
```

```python
### Inference:
# most pasengers were between the age of 15 to 40
```

```python
## 2. How does the survival rate vary based on passenger class (Pclass)?

sns.countplot(x='Survived', hue='Pclass', data=data, palette='bright')
plt.title('Survival by Passenger Class 0=Not Survived , 1= Survived')
plt.show()
```

```python
### Inference:
#Pclass 3 have highest count of perished passengers and Pclass 1 have highest survived rate
```

```python
## 3. Is there a relationship between the fare paid and the likelihood of survival?

sns.barplot(x='Survived', y='Fare', data=data)
plt.title('Survival by Fare')
plt.show()
```

```python
### Inference:
#Passengers who paid higher fare had higher chances of survival
```

```python
## 4. How does the survival rate differ between male and female passengers?

sns.countplot(x='Survived', hue='Sex', data=data, palette='bright')
plt.title('Survival by Sex')
plt.show()

```

```python
### Inference:
# Male passengers had lower rate of survival
```

```python
## 5. What impact does the number of siblings/spouses (SibSp) or parents/children(Parch) aboard have on survival?

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.countplot(x='SibSp', hue='Survived', data=data)
plt.title('Survival by SibSp')
plt.subplot(1,2,2)
sns.countplot(hue='Survived', x='Parch', data=data)
plt.title('Survival by Parch')
plt.show()
```

```python
### Inference:
# lower the count of siblings/spouses (SibSp) or parents/children(Parch) aboard have high survival rate
```

```python
## 6. Are there any significant differences in survival rates among passengers embarked from different ports?

sns.countplot(x='Survived', hue='Embarked', data=data, palette='bright')
plt.title('Survival by Embarked')
plt.show()
```

```python
### Inference:
# people embarked at S and Q
port had lower survival rate
```
