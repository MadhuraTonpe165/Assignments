```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import kagglehub
```

```python
#Download latest version
path = kagglehub.dataset_download("uciml/iris")

#print("Path to dataset files:", path)
```

```python
#name of the file
file_name = "Iris.csv"

# Construct the full file path
file_path = os.path.join(path, file_name)

# load data from csv file
df = pd.read_csv(file_path)
```

```python
#Display the first few rows of the dataset
df.head()
```

```python
#1. What is the distribution of sepal lengths and widths among the Iris species?

#using replot to plot graphs of length vs width for each species separately
sns.relplot(data=df, x='SepalLengthCm',y='SepalWidthCm', col='Species', palette='bright')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
```

```python
### Inference:
#Distribution of Width for
# Iris-setosa - 2.3cm to 4.4cm
# Iris-versicolor 2cm to 3.4cm
# Iris-Virginica 2.2cm to 3.8cm

#Distribution of Length for
# Iris-setosa - 4.1cm to 5.6cm
# Iris-versicolor 4.8cm to 8cm
# Iris-Virginica 4.8cm to 7.1cm
```

```python
# Q2. How does the petal length vary across different Iris species?

plt.figure(figsize=(15, 10))
sns.countplot(x='PetalLengthCm', hue='Species', data=df, palette='bright')
plt.title('Petal Length vs Species')
plt.ylabel('count')
plt.xlabel('PetalLengthCm')
plt.show()
```

```python
### Inference:
#Petal length for Iris-setosa is the smallest compared with others.
#Petal lenth for Iris-virginica is larget. Petal length for Iris-versicolor lies in middle
```

```python
#3. Is there a relationship between sepal length and petal length?
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.stripplot(y='SepalLengthCm',x='Species', data=df)
plt.xlabel('SepalLengthCm')
plt.ylabel('Species')
plt.subplot(1,2,2)
sns.stripplot(y='PetalLengthCm',x='Species', data=df)
plt.xlabel('PetalLengthCm')
plt.ylabel('Species')
plt.show()
```

```python
### Inference:
#when compared all species , Iris-setosa have smallest values of Sepal length and Petal length. similerly Iris-virginica have biggest Sepal lnegth and Petal length values
```

```python
#4. How does the petal width vary across different Iris species?

sns.stripplot(y='PetalWidthCm', x='Species', data=df)
plt.ylabel('PetalWidthCm')
plt.xlabel('Species')
plt.show()
```

```python
### Inference:
#When compared Iris-setosa have smallest petal width and Iris-virginica have biggest petal width
```

```python
#Q6. Can we distinguish between Iris species based on sepal and petal measurements?

sns.relplot(data=df, x='SepalLengthCm',y='SepalWidthCm', col='Species', palette='bright')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
sns.relplot(data=df, x='PetalLengthCm',y='PetalWidthCm', col='Species', palette='bright')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.show()
```

```python
### Inference:
#Iris-Setosa have smallest values of Sepal length, sepal width and petal length, Petal width
#Iris- Virginica have biggest alue set of Sepal length, sepal width and petal length, Petal width
#Iris-Versicolor Sepal length, sepal width and petal length, Petal width values range from smalll to large
```
