```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import os
import kagglehub
import numpy as np
```

```python
#Download latest version
path = kagglehub.dataset_download("uciml/iris")

print("Path to dataset files:", path)
```

```python
#name of the file
file_name = "Iris.csv"

# Construct the full file path
file_path = os.path.join(path, file_name)

# load data from csv file
df = pd.read_csv(file_path)
```

```python
#Display the first few rows of the dataset
df.head()
```

```python
# Question 1. What is the distribution of sepal lengths and widths among the Iris species?

plt.figure(figsize = (10,5))
'''
plt.scatter(x=df.SepalLengthCm, y=df.SepalWidthCm) gives values for all species
  but do not highlight them by species as scatter do not have hue argument.
'''
sns.scatterplot(x=df.SepalLengthCm, y=df.SepalWidthCm, hue=df.Species,  palette='viridis', s=100)
plt.title('Distribution of Sepal Length and Width by Iris Species')
plt.show()
```

```python
'''
#Histplot
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True, palette='pastel', element='step')
plt.title('Distribution of Sepal Length by Species')
plt.show()

plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='SepalWidthCm', hue='Species', kde=True, palette='pastel',element='step')
plt.title('Distribution of Sepal Width by Species')
plt.show()

#Pairplot - displays all cloums distribution with each other
sns.pairplot(df, hue='species', palette='viridis')
plt.show()
'''

#Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(y=df.SepalLengthCm, hue=df.Species, palette='viridis')

plt.figure(figsize=(10, 5))
sns.boxplot(y=df.SepalWidthCm, hue=df.Species, palette='viridis')
```

```python
'''
Inference:
 1. If we want to cluster data points by species, we can easily create a cluster for Iris-Setosa
 2. Iris-virginica data have few outliers for both SepalLengthCm and SepalWidthCm
'''
```

```python
# 2. How does the petal length vary across different Iris species?
```

```python
# Setting the figure size
plt.figure(figsize = (5, 3))

# Plot
plt.bar(df.Species, df.PetalLengthCm, color = 'black', width = 0.6)

# Set the x and y labels
plt.xlabel('Species')
plt.ylabel('PetalLengthCm')
plt.title('Petal Length by Species')
plt.show()
```

```python
'''
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, hue='Species', y='PetalLengthCm', palette='viridis')
plt.title('Petal Length Variation Across Iris Species')
plt.show()
'''
```

```python
# Setting the figure size
plt.figure(figsize=(8, 6))

#plot
sns.histplot(data=df, x='PetalLengthCm', hue='Species', kde=True, palette='viridis', element='step')
plt.title('Histogram of Petal Length by Species')
plt.show()
```

```python
'''
Inference:
Iris setosa: Typically has shorter petal lengths.
Iris versicolor: Has medium petal lengths with slight overlap with other species.
Iris virginica: Has the longest petal lengths on average
'''
```

```python
# 3. Is there a relationship between sepal length and petal length?
```

```python
# Setting the figure size
plt.figure(figsize = (15, 10))
plt.title('Relationship Between Sepal Length and Petal Length by Species')
plt.subplot(1,2,1)
sns.stripplot(data=df, x='Species', y='PetalLengthCm', palette='viridis')
plt.ylabel('PetalLengthCm')
plt.xlabel('Species')
plt.subplot(1,2,2)
sns.stripplot(data=df, x='Species', y='SepalLengthCm', palette='viridis')
plt.ylabel('SepalLengthCm')
plt.xlabel('Species')

plt.show()
```

```python
'''
Inference:
when compared all species , Iris-setosa have smallest values of Sepal length and Petal length. similerly Iris-virginica have biggest Sepal lnegth and Petal length values
'''
```

```python
# 4. How does the petal width vary across different Iris species?
```

```python
#setting the figure size
plt.figure(figsize = (5,3))

sns.boxplot(data=df, x='Species', y='PetalWidthCm', palette='viridis')
plt.title('Petal Width Variation Across Iris Species')
plt.show()

```

```python
#setting the figure size
plt.figure(figsize=(8, 6))

#Plot
sns.histplot(data=df, x='PetalWidthCm', hue='Species', kde=True, palette='viridis', element='step')
plt.title('Histogram of Petal Width by Species')
plt.show()
```

```python
'''
Inference :
Iris setosa: Generally has the smallest petal width.
Iris versicolor: Has a moderate petal width, with some overlap with other species.
Iris virginica: Has the largest petal width on average, with more variation compared to setosa.
'''
```

```python
# 5. Are there any patterns observable in the pairwise relationships between features?
```

```python
#Pairplot - displays all cloums distribution with each other
sns.pairplot(df, hue='Species', palette='viridis')
plt.show()
```

```python
'''
Inference:
 1. Features like petal length and petal width are highly correlated and effectively differentiate species.
 2. Iris setosa stands out as the most distinct species, while versicolor and virginica show some overlap in their feature values.
'''
```

```python
# 6. Can we distinguish between Iris species based on sepal and petal measurements?
```

```python
#setting the figure size
plt.figure(figsize=(8, 6))

#Plot
sns.displot(x=df.PetalLengthCm, y=df.PetalWidthCm, hue=df.Species,)
sns.displot(x=df.SepalLengthCm, y=df.SepalWidthCm, hue=df.Species,)
plt.plot()
```

```python
'''
Inference:
1. Setosa can be easily distinguished from Versicolour and Virginica.
2. from Sepalwidth and Sepal length values Versicolor and Virginica can not be identified. Petal Width and Petal length might help but still few values are overlapping
'''
```
