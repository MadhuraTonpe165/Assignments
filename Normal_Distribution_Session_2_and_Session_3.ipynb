```python
#Importing Important Libraries
import numpy as np
import pandas as pd
```

```python
#Importing the 'BEML' dataset
beml_df=pd.read_csv('BEML.csv')
beml_df.head()
```

```python
#Importing the 'Glaxo' dataset
glaxo_df=pd.read_csv('GLAXO.csv')
glaxo_df.head()
```

```python
# Filtering the required columns
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
# Getting the data types for BEML
beml_df.dtypes
```

```python
# Getting the data types for Glaxo
glaxo_df.dtypes
```

```python
# Creating the Index for BEML
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
beml_df.head()
```

```python
# Creating Index for Glaxo
glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
glaxo_df.head()
```

```python
#  Data Visualization

# Import Libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot for Glaxo
plt.figure(figsize = (5, 3))
plt.plot(glaxo_df.index, glaxo_df.Close)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Closing Proce of Glaxo')
plt.show()
```

```python
# Plot for BEML
plt.figure(figsize = (5, 3))
plt.plot(beml_df.index, beml_df.Close)
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title('Closing Proce of BEML')
plt.show()
```

```python
# Adding a 'gain' column for glaxo
glaxo_df['gain'] = glaxo_df.Close.pct_change(periods = 1) # pct_change is a general fucntion not related to any library
glaxo_df.head()
```

```python
# Adding a 'gain' column for beml
beml_df['gain'] = beml_df.Close.pct_change(periods = 1)
beml_df.head()
```

```python
# Dropping the row with NAN value, drop nan and null values
glaxo_df.dropna(inplace = True)
beml_df.dropna(inplace = True)
```

```python
# check the values
glaxo_df.head()
beml_df.head()
```

```python
# plot the gains
plt.figure(figsize = (5, 3))
plt.plot(beml_df.index, beml_df.gain)
plt.xlabel('year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```

```python
plt.figure(figsize = (5, 3))
plt.plot(glaxo_df.index, glaxo_df.gain)
plt.xlabel('year')
plt.ylabel('Gain')
plt.title('Gain of Glaxo')
plt.show()
```

```python
# Plot both gain in same graph

plt.figure(figsize = (5, 3))
plt.plot(beml_df.index, beml_df.gain, c = 'red')
plt.plot(glaxo_df.index, glaxo_df.gain, c = 'blue')
plt.xlabel('year')
plt.ylabel('Gain')
plt.title('Gain')
plt.show()
```

```python
#Getting the Statictical Properties
print('Mean :', round(glaxo_df['gain'].mean(), 4))
print('Standard Deviation :', round(glaxo_df['gain'].std(), 4))
```

```python
#Getting the Statictical Properties
print('Mean :' , round(beml_df['gain'].mean(), 4))
print('Standard Deviation :' , round(beml_df['gain'].std(), 4))
```

```python
# Getting the 95% confidence interval
from scipy import stats
glaxo_df_ci = stats.norm.interval(0.95,
                                  loc= glaxo_df.gain.mean(),
                                  scale=glaxo_df.gain.std())

#Confidence Interval
print('Glaxo Confidence Interval :' , np.round(glaxo_df_ci, 4))
```

```python
# Getting the 95% confidence interval for BEML

beml_df_ci= stats.norm.interval(0.95,
                                loc=beml_df.gain.mean(),
                                scale=beml_df.gain.std())
#Confidence Interval
print('BEML Confidence Interval :' , np.round(beml_df_ci, 4))
```
