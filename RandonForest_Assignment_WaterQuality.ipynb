```python
# Importing Important Libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mssmartypants/water-quality")

print("Path to dataset files:", path)
```

```python
# Load the data
df = pd.read_csv(f'{path}/waterQuality1.csv')
#/root/.cache/kagglehub/datasets/mssmartypants/water-quality/versions/3/waterQuality1.csv

# Show first 5 rows of the data
df.head()
```

```python
df.info()
```

```python
df['ammonia'].value_counts()
```

```python
# Few values in ammonia column have value as '#NUM!'
# make ammonia -> numeric
df.loc[df['ammonia'] == '#NUM!', 'ammonia']= df['ammonia'].mode()[0]
```

```python
# converting data type of ammonia column to float
df['ammonia']= df['ammonia'].astype(float)
```

```python
# Few roew in is_safe column have value as '#NUM!'
# make is_safe -> numeric
df.loc[df['is_safe'] == '#NUM!', 'is_safe']= df['is_safe'].mode()[0]
df['is_safe']= df['is_safe'].astype(int)
```

```python
df['is_safe'].value_counts()
```

```python
# Separating features and target
X = df.drop(['is_safe'], axis = 1)
y = df['is_safe']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 5)
```

```python
# Creating a Random Forest Classifier
clf = RandomForestClassifier(n_estimators= 200)

# Fitting the data
clf.fit(X_train, y_train)
```

```python
# Taking the predictions from the model

y_pred = clf.predict(X_test)
y_train_pred = clf.predict(X_train)
```

```python
# Model Evaluation
print('Training Accuracy :', np.round(metrics.accuracy_score(y_train, y_train_pred), 2))
print('Training Precision :', np.round(metrics.precision_score(y_train, y_train_pred), 2))
print('Training Recall :', np.round(metrics.recall_score(y_train, y_train_pred), 2))
print('Training F1 Score :', np.round(metrics.f1_score(y_train, y_train_pred), 2))

print('\nTesting Accuracy :', np.round(metrics.accuracy_score(y_test, y_pred), 2))
print('Testing Precision :', np.round(metrics.precision_score(y_test, y_pred), 2))
print('Testing Recall :', np.round(metrics.recall_score(y_test, y_pred), 2))
print('Testing F1 Score :', np.round(metrics.f1_score(y_test, y_pred), 2))
```
