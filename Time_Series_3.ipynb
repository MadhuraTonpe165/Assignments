```python
# Importing Important Lbraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Creating a sample time series data
date_range = pd.date_range(start = '1/1/2020' , periods = 100, freq = 'D')
data = np.random.randn(100)
time_series = pd.Series(data, index = date_range)
time_series.head()
```

```python
from datetime import time
# Dealing with missing values

time_series_with_nan = time_series.copy()
time_series_with_nan[::10] = np.nan
time_series_with_nan.head(50)
```

```python
# Forward Filling Method
time_series_forward = time_series_with_nan.fillna(method = 'ffill')
time_series_forward.head(50)
```
